```python
import cv2
import numpy as np
from matplotlib import pyplot as plot
from yolo import yolo
%matplotlib inline
```

```python
video_file = '/home/amin/Documents/raspbery/yolo/video_2018-12-07_00-12-18.mp4' # contains 4617 frames
tmp_img = '/tmp/tmp.jpg'
```

```python
y = yolo()
cap = cv2.VideoCapture(video_file)
i = 1
while True:
    _, frame = cap.read()
    if frame is None:
        break
#    img = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
#    fig2=plot.figure(figsize=[5,5])
#    a = plot.imshow(img)
    i += 1
    if i % 1100 == 0:
        cv2.imwrite(tmp_img, frame)
        plot.figure()
        y.set_image(tmp_img,3)
        print('--------------------------')
cap.release()
```

I know it is not appropriate to save the frame, but it's the fastest way to implement :D
